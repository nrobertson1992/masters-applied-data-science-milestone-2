```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

# Merge County level data

```python
poverty_df = pd.read_csv('CLEANED county_poverty BACKUP.csv',index_col=False)
education_df = pd.read_csv('CLEANED county_education BACKUP.csv',index_col=False)
unemployment_df = pd.read_csv('CLEANED county_unemployment BACKUP.csv',index_col=False)
```

```python
poverty_df.drop(columns=['Unnamed: 0'],inplace=True)
poverty_df['joincol'] = poverty_df['county_name'] + ', ' + poverty_df['state']
poverty_df.head()
```

```python
education_df.drop(columns=['Unnamed: 0'],inplace=True)
education_df['joincol'] = education_df['county_name'] + ', ' + education_df['state']
education_df.head()
```

```python
unemployment_df.drop(columns=['Unnamed: 0'],inplace=True)
unemployment_df['joincol'] = unemployment_df['county_name'] + ', ' + unemployment_df['state']
unemployment_df.head()
```

```python
county_df = poverty_df.merge(
    education_df[['joincol','perc-college-educated-2016-2020']],
    on='joincol',
    how='inner'
)
county_df = county_df.merge(
    unemployment_df[['joincol','perc_unemployed_2021','median_hh_income','perc_state_median_hh_income']],
    on='joincol',
    how='inner'
)

county_df = county_df[['county_name','state','FIPS','rural-urban-continuum-code',
                       'perc-college-educated-2016-2020','perc_unemployed_2021','median_hh_income',
                       'perc_state_median_hh_income','perc_people_poverty_2020','perc_children_poverty_2020']]

county_df.head()
```

```python
county_df.to_csv('CLEANED county_data.csv')
```

```python
scraped_listings = pd.read_csv('CLEANED scraped_listings BACKUP.csv')
scraped_pages = pd.read_csv('CLEANED scraped_pages BACKUP.csv')
geographic_data = pd.read_csv('CLEANED geolocation_data.csv')
```

```python
scraped_listings['city'] = scraped_listings['listing_city']
scraped_listings.drop(columns=['Unnamed: 0'],inplace=True)
scraped_listings.head()
```

```python
scraped_pages.drop(columns=['Unnamed: 0'],inplace=True)
scraped_pages['listing_id'] = scraped_pages['listing_id'].astype('int',errors = 'ignore')
scraped_pages.head()
```

```python
geographic_data.drop(columns=['Unnamed: 0','latitude','longitude'],inplace=True)
geographic_data.head()
```

```python
airbnb_data = scraped_listings.merge(
    scraped_pages, 
    how='left',
    on='listing_id'
)

airbnb_data = airbnb_data.merge(
    geographic_data,
    how='left',
    on='listing_id'
)

airbnb_data.head()
```

```python
airbnb_data = airbnb_data[['listing_id','listing_url','is_superhost','is_premium','rating',
                          'n_reviews','country','state','city','county','postcode','listing_title','n_pictures','entire_home',
                         'hotel_room','private_room','shared_room','latitude','longitude','price',
                         'perc_discount','n_guests','n_bedrooms','n_beds','n_baths',
                         'rating_cleanliness','rating_communication','rating_checkin','rating_accuracy',
                         'rating_location','rating_value','n_amenities','amenities_services','amenities_heat_cool',
                         'amenities_bedroom_laundry','amenities_kitchen_dining','amenities_safety',
                          'amenities_parking_facilities','amenities_internet_office','amenities_not_included',
                          'amenities_entertainment','amenities_bathroom','amenities_outdoor',
                          'amenities_location_features','amenities_family','amenities_scenic_views','bed_king',
                          'bed_queen','bed_double','bed_single','bed_sofa','bed_air','bed_floor','bed_bunkbed',
                          'bed_hammock','bed_baby','description','description_sentiment_score','description_length',
                          'description_reading_level','image_1','image_2','image_3','image_4',
                          'image_5'
                         ]]


airbnb_data.head(10)
```

```python
def clean_rating(x):
    try:
        if x == 'New':
            return None
        else:
            return x
    except:
        return x
    
airbnb_data['rating'] = airbnb_data['rating'].apply(clean_rating)
airbnb_data['n_reviews'] = airbnb_data['n_reviews'].apply(clean_rating)


def clean_stats(x):
    try:
        if '+' in x:
            return int(x.replace('+',''))
        else:
            return x
    except:
        return x
    
    
airbnb_data['n_guests'] = airbnb_data['n_guests'].apply(clean_stats)
airbnb_data['n_bedrooms'] = airbnb_data['n_bedrooms'].apply(clean_stats)
airbnb_data['n_beds'] = airbnb_data['n_beds'].apply(clean_stats)
airbnb_data['n_baths'] = airbnb_data['n_baths'].apply(clean_stats)
    
def studio(x):
    try:
        if x=='Studio':
            return 0
        else:
            return x
    except:
        return x
    
airbnb_data['n_bedrooms'] = airbnb_data['n_bedrooms'].apply(studio)


def halfbath(x):
    try:
        if x=='Half-bath':
            return 0
        else:
            return x
    except:
        return x
    
airbnb_data['n_guests'] = airbnb_data['n_guests'].apply(halfbath)
airbnb_data['n_bedrooms'] = airbnb_data['n_bedrooms'].apply(halfbath)
airbnb_data['n_beds'] = airbnb_data['n_beds'].apply(halfbath)
airbnb_data['n_baths'] = airbnb_data['n_baths'].apply(halfbath)

def shared(x):
    try:
        if x=='Shared':
            return 0
        elif x=='Private':
            return 1
        else:
            return x
    except:
        return x
    
airbnb_data['n_beds'] = airbnb_data['n_beds'].apply(shared)
airbnb_data['n_baths'] = airbnb_data['n_baths'].apply(shared)
```

# Remove listings in areas we won't analyze.

Due to the nature of how the scraping worked, there were some listings at the edges of the geographic areas we want to assess that were scrapped. This brought into data from border states (Nebraska), and border countries (Canada), that share a border with a state we scrapped.

We'll drop these listings, as we don't have the relevant demographic data about the area to help assess those listings in the same way we will assess the others.

```python
airbnb_data = airbnb_data[~airbnb_data['country'].isin(['Canada','Mxico'])]
```

```python
airbnb_data = airbnb_data[~airbnb_data['state'].isin(['Nebraska','North Dakota','Texas','South Dakota','Oklahoma'])]
```

```python
airbnb_data.to_csv('CLEANED airbnb_data.csv')
```
